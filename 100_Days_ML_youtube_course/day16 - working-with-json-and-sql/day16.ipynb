```python
import subprocess
from pathlib import Path

sql_path = Path("world.sql")

# import into MySQL (will run whatever is inside world.sql)
# If world.sql contains CREATE DATABASE world; USE world; it will set itself up.
subprocess.run(
    ["mysql", "-u", "root", "--socket=/tmp/mysql.sock"],
    input=sql_path.read_bytes(),
    check=True
)

print("Imported world.sql successfully")
```

```python
import pandas as pd
import mysql.connector

conn = mysql.connector.connect(
    user="root",
    password="",
    database="world",
    unix_socket="/tmp/mysql.sock"
)

print("connected:", conn.is_connected())

df = pd.read_sql("SELECT * FROM countrylanguage WHERE Language LIKE '%Spanish%'", conn)
df.head()
```
